```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('/Users/puravgupta/Desktop/private/WhatsApp Image 2024-06-28 at 19.02.43.jpeg', cv2.IMREAD_GRAYSCALE)  
blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
image = cv2.imread('C:\Users\utkar\Desktop')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  

faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5)

for (x, y, w, h) in faces:

    face_region = gray_image[y:y+h, x:x+w]
    blurred_face = cv2.GaussianBlur(face_region, (5, 5), 1.4)
    edges = cv2.Canny(blurred_face, threshold1=100, threshold2=200)
    image[y:y+h, x:x+w] = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)


plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Face Detection with Canny Edge Detection')
plt.axis('off')
plt.show()
```
